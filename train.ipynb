```python
import pandas as pd 
import numpy as np
```

```python
# read text corpus 

def read_corpus(file_path):
    with open(file_path, 'r') as file:
        corpus = file.read()
    return corpus

# read the corpus
corpus = read_corpus('data/wiki2.train.txt')

# print the first 1000 characters of the corpus
print(corpus[:1000])
```

```python
# build the vocabulary

def build_vocabulary(corpus):
    # split the corpus into words
    words = corpus.split()
    # remove duplicate words
    vocabulary = list(set(words))
    return vocabulary

# build the vocabulary
vocabulary = build_vocabulary(corpus)

#check the size of the vocabulary
print(len(vocabulary))
```

```python

# reduce the vocab size using lemmatization

from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')

def lemmatize_vocabulary(vocabulary):
    lemmatizer = WordNetLemmatizer()
    lemm_vocab = set()
    for word in vocabulary:
        lemm_word = lemmatizer.lemmatize(word)
        lemm_vocab.add(lemm_word)
    return lemm_vocab

# lemmatize the vocabulary
lemmatized_vocabulary = lemmatize_vocabulary(vocabulary)

# check the size of the lemmatized vocabulary
print(len(lemmatized_vocabulary))
```

```python
# create a frequency count of words

def get_word_frequency(corpus):
    words = corpus.split()
    word_freq = {}
    for word in words:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

    
    # sort the dictionary by value
    word_freq = dict(sorted(word_freq.items(), key=lambda item: item[1], reverse=True))
    return word_freq

# get the word frequency
word_freq = get_word_frequency(corpus)

# print the 10000 most frequent words
print(dict(list(word_freq.items())[:10000]))
```
